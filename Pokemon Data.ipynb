# Data Analysis w/ Pokemon Data
## Step 1 : Import NumPy, MatPlotLib, SKLearn

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Step 2 : Read Data from csv file

```python
data = pd.read_csv('Pokemon.csv')
data
```

```python
#name = data['Name']
pokemon = data[['Name', 'Type 1', 'Attack', 'Legendary']]
#pokemon
```

## Step 3: Print out amount of type

```python
## Number of types
types = data['Type 1'].value_counts()
#type
```

## Step 4: Making Plots

```python
## Horizontal Bar Graph based off Type 1 and Attack
plt.figure(figsize=(20, 20) , facecolor='white')
attack_graph = plt.barh(pokemon['Type 1'], pokemon['Attack'], color= "green")
## add Labels to show the Type, Attack Power, etc.
plt.ylabel("Type")
plt.xlabel("Attack Power")
plt.title("Amount of Attack Power based off Type")
```

```python
plt.figure(figsize=(30, 30) , facecolor='white')
plt.scatter(data['Defense'], pokemon['Attack'])
plt.ylabel("Defense")
plt.xlabel("Attack Power")
plt.title("Amount of Defense compared to Attack")

for i, txt in enumerate(data['Name']):
    plt.annotate(txt, (data.Attack[i], data.Defense[i]))
```
